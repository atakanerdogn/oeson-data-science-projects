```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

x = casino_numeric.drop('Total Gross Gaming Revenue', axis = 1)
y = casino_numeric['Total Gross Gaming Revenue']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

L = LinearRegression()
L.fit(x_train, y_train)

y_pred = L.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)

df = pd.DataFrame({'y_test': y_test, 'y_pred': y_pred})
df.head()
```

```python
sns.lineplot(x='y_test', y='y_pred', data=df)
plt.title('Linear Regression')
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)

y_pred_rf = rf.predict(x_test)

mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print("Mean Squared Error (Random Forest):", mse_rf)
print("R-squared (Random Forest):", r2_rf)

df_rf = pd.DataFrame({'y_test': y_test, 'y_pred_rf': y_pred_rf})
df_rf.head()

sns.lineplot(x='y_test', y='y_pred_rf', data=df_rf)
plt.title('Random Forest')
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gb = GradientBoostingRegressor(n_estimators=100, random_state=42)
gb.fit(x_train, y_train)

y_pred_gb = gb.predict(x_test)

mse_gb = mean_squared_error(y_test, y_pred_gb)
r2_gb = r2_score(y_test, y_pred_gb)

print("Mean Squared Error (Gradient Boosting):", mse_gb)
print("R-squared (Gradient Boosting):", r2_gb)

df_gb = pd.DataFrame({'y_test': y_test, 'y_pred_gb': y_pred_gb})
df_gb.head()

sns.lineplot(x='y_test', y='y_pred_gb', data=df_gb)
plt.title('Gradient Boosting')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(random_state=42)
dt.fit(x_train, y_train)

y_pred_dt = dt.predict(x_test)

mse_dt = mean_squared_error(y_test, y_pred_dt)
r2_dt = r2_score(y_test, y_pred_dt)

print("Mean Squared Error (Decision Tree):", mse_dt)
print("R-squared (Decision Tree):", r2_dt)

df_dt = pd.DataFrame({'y_test': y_test, 'y_pred_dt': y_pred_dt})
df_dt.head()

sns.lineplot(x='y_test', y='y_pred_dt', data=df_dt)
plt.title('Decision Tree')
plt.show()
```
